## üöÄ High_Performance_1D_Finite_Element_Method
This code implements a 1D finite elements method adapted from Pavel Solin's book "Partial Differential Equations and the Finite Element Method."

By first creating an element class and using sparse matrix functionality, we can vary the maximal polynomial degree of Lobatto shape functions, along with the required number of quadrature points, on each element. 

## üî¨ Model Problem

We solve:

$$
- \nabla \cdot (a_1 \nabla u) + a_0 u = f
$$

where:

- $ f \in L^2(\Omega) $
- $ a_1, a_0 \in L^\infty(\Omega) $
- $ a_1(x) \geq C_{min} > 0 $
- $ a_0(x) \geq 0 $ in $ \Omega $.

---

## üí° Weak Formulation

Find \( u \in V \) such that:

$$
a(u,v) = \ell(v) \quad \text{for all} \quad v \in V
$$

where

$$
a(u,v)
=
\int_\Omega \left[ a_1 \nabla u \cdot \nabla v + a_0 u v \right] \, dx,
$$

and

$$
\ell(v)
=
\int_\Omega f v \, dx.
$$

---

## üìù Galerkin Discretization

For basis functions $ \{v_i\}_{i=1}^N $:

- **Stiffness matrix entries**:

$$
s_{ij}
=
a(v_j, v_i),
$$

for $ i,j = 1,\ldots,N $, forming $ S $.

- **Load vector entries**:

$$
f_i
=
\ell(v_i),
$$

for $ i = 1,\ldots,N $, forming $ F $.

We solve:

$$
S Y = F
$$

where $ Y $ is the vector of solution coefficients.

---

## ‚öôÔ∏è Element Class Procedure

1. **Select domain**: $ \Omega = (-1,1) $ in 1D.
2. **Affine mapping**: For each element $ K_i $, construct an affine map from reference element $ \hat{K} $ to $ K_i $.
3. **Transform weak forms** to the reference element.
4. **Define shape functions** in $ \hat{K} $.
5. **Build global basis** by mapping to physical elements.
6. **Enumerate shape functions** to build the connectivity array linking local and global DOFs.
7. **Assemble** element by element to solve $ S Y = F $.
